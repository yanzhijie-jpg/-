```python
import tensorflow as tf 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
#加载用到的工具
data = pd.read_csv('yan.csv',engine='python') 
data1= pd.read_csv('yan.csv',engine='python') 
```

```python
data
```

```python
x=data.iloc[:45,2: ] 
y=data.iloc[:45,1:2] 
z=data1.iloc[45:-4, 2:] 
w=data1.iloc[45:-4 ,1:2] 
```

```python
w_1=w.copy()
```

```python
w_1
```

```python
x= (x-x.min())/(x.max()-x.min()) 
y=(y-y.min())/(y.max()-y.min()) 
z=(z-z.min())/(z.max()-z.min())
w=(w-w.min())/(w.max()-w.min())
```

```python
print(x.shape )
print(y.shape)
y
```

```python
#定义模型
model = tf.keras.Sequential() 
model.add(tf.keras.layers.Dense(200,input_shape=(16,), activation='relu'))
 #输入层限定输入数 据规模，设置 Dense 层，激活函数为 relu 
#model.add(tf.keras.layers.Dropout(0.2))
 #隐藏层 Dropout 层 
model.add(tf.keras.layers.Dense(1))
 #输出层，限定输出数据规模，激活函 数为 softmax 
    

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='mae')
```

```python
model.summary()
```

```python
history=model.fit(x,y,epochs=100,validation_data=(z,w))
```

```python
#绘制损失拟合图
plt.plot(history.epoch,history.history.get('loss'), label='loss') 
plt.plot(history.epoch,history.history.get('val_loss'),label='val_loss') 
plt.legend()
```

```python
pre=model.predict(z)
```

```python
#计算mse损失
def mse(pre,w):
    w=np.array(w)
    dis=[]
    for  i  in range(w.shape[0]):
        
        val=w[i][0]-pre[i][0]
        dis.append(val**2)
    result=np.cumsum(dis,axis=0)[-1]/len(dis)
    return result
```

```python
result=mse(pre,w)
print('bp_mse:{} gen_loss is {},} '.format(result,
                                                        ))#打印当前次数，及损失函数情况
```
